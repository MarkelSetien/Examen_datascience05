```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
import requests
from lxml import html
```

```python
df = pd.read_csv("C:/Users/Markel/Documents/BBDD2/datascience/reto06/Examen/netflix_titles (1).csv", sep=',')
```

```python
df.head()
```

## Recomendador de películas (2,5 puntos)

```python
df_movies=pd.DataFrame(df[df['type']=='Movie'])
```

```python
print(df_movies)
```

```python
df_movies=df_movies.reset_index()
```

```python
df_movies2=df_movies.reset_index()
```

```python
df_movies2
```

```python
df_movies.isna()
```

```python
df_movies.isna().sum()
```

```python
df_movies['director'] = df_movies['director'].fillna('')
df_movies['cast'] = df_movies['cast'].fillna('')
df_movies['country'] = df_movies['country'].fillna('')
df_movies['rating'] = df_movies['rating'].fillna('')
df_movies['duration'] = df_movies['duration'].fillna('')
```

```python
df_movies.isnull().sum()
```

```python
def combined(row):
    return str(row['director'])+" "+str(row['cast'])+" "+str(row['listed_in'])+" "+str(row['description'])
df_movies["combined"] = df_movies.apply(combined, axis =1)
```

```python
df_movies['combined']
```

```python
vectorizer = CountVectorizer().fit_transform(df_movies["combined"].values)
vectors = vectorizer.toarray()
csim = cosine_similarity(vectors)
```

```python
csim
```

### Escrapeo de la película

```python
page= requests.get('https://dreguera.github.io/')
tree= html.fromstring(page.content)
```

```python
tree
```

```python
a=tree.xpath('//div[@class="morado"]/text()')
a
```

```python
pelicula=a[0]
```

```python
serie=a[1]
```

```python
def get_index_from_title(title):
    return df_movies2[df_movies2.title == title]["index"].values[0]
```

```python
index = get_index_from_title(pelicula)
```

```python
index= get_index_from_title('Young Tiger')
index
```

```python
def get_level_0_from_title(title):
    return df_movies2[df_movies2.title == title]["level_0"].values[0]
```

```python
level=get_level_0_from_title(pelicula)
level
```

```python
def get_title_from_index(index):
    return df_movies[df_movies.index == index]["title"].values[0]
```

```python
similar_movies = list(enumerate(csim[level]))
```

```python
similar_movies
```

```python
sorted_similar_movies = sorted(similar_movies, key=lambda x:x[1], reverse=True)
```

```python
sorted_similar_movies
```

```python
def get_title_from_index(index):
    return df_movies[df_movies.index == index]["title"].values[0]

contador = 0 
for movie in sorted_similar_movies: 
    print(get_title_from_index(movie[0]))
    contador += 1
    if contador == 15:
        break
```
